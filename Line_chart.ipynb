```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = {
    'x': np.random.rand(10) * 10,
    'y': np.random.rand(10) * 10,
    'z': np.random.rand(10) * 100
}
df = pd.DataFrame(data)
```

```python
#Informações básicas sobre o DataFrame
#df = pd.read_csv('Caminho relativo')
df.info()
print(f'-----------')
print(f'Número de colunas: {len(df.columns)}')
print(f'-----------')
print(f'Nomes das colunas: {df.columns}')
```

```python
#Opções de amostras do data frame.
#tamanho limite da visualização das amostras.
pd.set_option('display.max_columns', 10) #None para sem limite
pd.set_option('display.max_rows', 20)
display(df)
```

Standart line plot

```python
# Ordena o DataFrame pela coluna 'x', para melhor vizualização do grafico de linha
df_sorted = df.sort_values(by='x').reset_index(drop=True)

# Configurações do Tamanho do Gráfico em centímetros
largura_cm = 18
altura_cm = 12
polegadas_por_cm = 1 / 2.54
largura_in = largura_cm * polegadas_por_cm
altura_in = altura_cm * polegadas_por_cm

# Configurações do Título e Rótulos
label_x = 'Variável X'
label_y = 'Variável Y'
tamanho_label = 14
cor_label = 'black'

#Configurações da Linha e Marcadores
cor_linha = 'purple'
espessura_linha = 2
tamanho_marcador = 5
cor_marcador = 'black' # Cor do preenchimento do ponto
cor_borda_marcador = 'gray' # Cor da borda do ponto

# Cria o gráfico com os dados ordenados e as personalizações.
plt.figure(figsize=(largura_in, altura_in))
plt.plot(df_sorted['x'], df_sorted['y'],
        marker='o',
        linestyle='-',
        color=cor_linha,
        label='Dados Ordenados',
        linewidth=espessura_linha,
        markersize=tamanho_marcador,
        markerfacecolor=cor_marcador,
        markeredgecolor=cor_borda_marcador
        )

plt.tick_params(axis='both', which='major', labelsize=10)
plt.xlabel(label_x, fontsize=tamanho_label, color=cor_label)
plt.ylabel(label_y, fontsize=tamanho_label, color=cor_label)

plt.grid(True)
plt.legend(fontsize=12)
plt.show()
#plt.savefig('grafico_linha.png', dpi=300)
```

2 line chart variations

```python
np.random.seed(42)
data = {
    'x': np.random.rand(20) * 10,
    'y': np.random.rand(20) * 10,
    'x1': np.random.rand(20) * 10,
    'y1': np.random.rand(20) * 10
}
df = pd.DataFrame(data)
df_sorted = df.sort_values(by='x').reset_index(drop=True)
df_sorted1 = df.sort_values(by='x1').reset_index(drop=True)

y_data = df['y']
y1_data = df['y1']
# Tamanho base (para 1 gráfico)
base_width_cm = 18
base_height_cm = 12
inches_per_cm = 1 / 2.54

# Tamanho total da figura
fig_width_in = (base_width_cm * 2) * inches_per_cm
fig_height_in = (base_height_cm) * inches_per_cm

# Configurações de Texto
tamanho_titulo = 16
cor_titulo = 'black'
tamanho_label = 14
cor_label = 'black'

# Configurações Plot 1 e 2 (linha)
cor_linha = 'purple'
espessura_linha = 2
tamanho_marcador = 5
cor_marcador = 'black'
cor_borda_marcador = 'gray'
#linha 2
cor_linha2 = 'red'
espessura_linha2 = 2
tamanho_marcador2 = 5
cor_marcador2 = 'black'
cor_borda_marcador2 = 'gray'


#Criação dos Subplots (Layout 1x2)
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(fig_width_in, fig_height_in))

#Plot 1: Gráfico de Linha (axs[0, 0])
ax1 = axs[0]
ax1.plot(df_sorted['x'], df_sorted['y'],
        marker='o',
        linestyle='-',
        color=cor_linha,
        label='Dados Ordenados',
        linewidth=espessura_linha,
        markersize=tamanho_marcador,
        markerfacecolor=cor_marcador,
        markeredgecolor=cor_borda_marcador)
ax1.tick_params(axis='both', which='major', labelsize=20)
ax1.set_xlabel('Variável X', fontsize=tamanho_label, color=cor_label)
ax1.set_ylabel('Variável Y', fontsize=tamanho_label, color=cor_label)
ax1.set_title('Relação entre as Variáveis X e Y', fontsize=tamanho_titulo, color=cor_titulo)
ax1.grid(True)
ax1.legend(fontsize=12)

#Plot 2: Gráfico de linha (Sem Legenda) (axs[0, 1])
ax2 = axs[1]
ax2.plot(df_sorted['x'], df_sorted['y'],
        marker='o',
        linestyle='-',
        color=cor_linha,
        linewidth=espessura_linha,
        markersize=tamanho_marcador,
        markerfacecolor=cor_marcador,
        markeredgecolor=cor_borda_marcador)
ax2.plot(df_sorted1['x1'], df_sorted1['y1'],
        marker='o',
        linestyle='-',
        color='red',
        linewidth=espessura_linha,
        markersize=tamanho_marcador,
        markerfacecolor=cor_marcador,
        markeredgecolor=cor_borda_marcador)
ax2.tick_params(axis='both', which='major', labelsize=20)
ax2.set_xlabel('Variável X', fontsize=tamanho_label, color=cor_label)
ax2.set_ylabel('Variável Y', fontsize=tamanho_label, color=cor_label)
ax2.set_title('Gráfico de Linha', fontsize=tamanho_titulo, color=cor_titulo)
ax2.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

#plt.savefig('line_subplots.png', dpi=300)
```
