```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#exemplo de DataFrame
np.random.seed(42)
data = {
    'x': np.random.rand(50) * 10,
    'y': np.random.rand(50) * 10,
    'z': np.random.rand(50) * 100
}
df = pd.DataFrame(data)
```

```python
#Informações básicas sobre o DataFrame
#df = pd.read_csv('Caminho relativo')
df.info()
print(f'-----------')
print(f'Número de colunas: {len(df.columns)}')
print(f'-----------')
print(f'Nomes das colunas: {df.columns}')
```

```python
#Opções de amostras do data frame.
#tamanho limite da visualização das amostras.
pd.set_option('display.max_columns', 30) #none para sem limite
pd.set_option('display.max_rows', 30)
display(df)
```

Standart Scatter plot

```python
# Tamanho da Figura em Centímetros
width_cm = 18
height_cm = 12
inches_per_cm = 1 / 2.54
width_in = width_cm * inches_per_cm
height_in = height_cm * inches_per_cm

# Título e Rótulos
titulo_do_grafico = 'Scatter Plot das Variáveis X e Y'
label_x = 'Variável X'
label_y = 'Variável Y'
tamanho_titulo = 16
cor_titulo = 'black'
tamanho_label = 14
cor_label = 'black'

# Configurações do Scatter Plot
tamanho_pontos = 150 # Tamanho dos marcadores
cor_cmap = 'viridis' # Mapa de cores
alpha_transparencia = 0.7 # Transparência dos pontos (0 a 1)
cor_borda_ponto = 'black'
espessura_borda_ponto = 0.8


#SCATTER PLOT SEM BARRA DE CORES
plt.figure(figsize=(width_in, height_in))
plt.scatter(df['x'], df['y'],
                    s=tamanho_pontos,
                    c=df['z'], # Cor dos pontos baseada na coluna 'z'
                    cmap=cor_cmap,
                    alpha=alpha_transparencia,
                    edgecolors=cor_borda_ponto,
                    linewidths=espessura_borda_ponto)

plt.xlabel(label_x, fontsize=tamanho_label, color=cor_label)
plt.ylabel(label_y, fontsize=tamanho_label, color=cor_label)
plt.title(f'{titulo_do_grafico} (Sem Legenda de Escala)', fontsize=tamanho_titulo, color=cor_titulo)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

#plt.savefig('scatter_plot_sem_legendade_escala.png', dpi=300)
```

Scatter-line chart

```python
# Tamanho da Figura em Centímetros
width_cm = 18
height_cm = 12
inches_per_cm = 1 / 2.54
width_in = width_cm * inches_per_cm
height_in = height_cm * inches_per_cm

# Título e Rótulos
titulo_do_grafico = 'Scatter Plot com Linha de Regressão'
label_x = 'Variável X'
label_y = 'Variável Y'
tamanho_titulo = 16
cor_titulo = 'black'
tamanho_label = 14
cor_label = 'black'

# Configurações do Scatter Plot
tamanho_pontos = 100
cor_cmap = 'viridis'
alpha_transparencia = 0.7
cor_borda_ponto = 'black'
espessura_borda_ponto = 0.8

#Configurações da linha de regressão
cor_linha_regressao = 'blue' # Cor da linha de regressão
estilo_linha_regressao = '-' # 'solid'
espessura_linha_regressao = 2

#SCATTER PLOT SEM BARRA DE CORES
plt.figure(figsize=(width_in, height_in))
plt.scatter(df['x'], df['y'],
                s=tamanho_pontos,
                c=df['z'], # Cor dos pontos baseada na coluna 'z'
                cmap=cor_cmap,
                alpha=alpha_transparencia,
                edgecolors=cor_borda_ponto,
                linewidths=espessura_borda_ponto)

#Calculando e plotando uma linha de regressão linear com base nos dados.
coefs = np.polyfit(df['x'], df['y'], 1)
p = np.poly1d(coefs)
x_line = np.linspace(df['x'].min(), df['x'].max(), 100)
y_line = p(x_line)

plt.plot(x_line, y_line,
        color=cor_linha_regressao,
        linestyle=estilo_linha_regressao,
        linewidth=espessura_linha_regressao,
        label='Linha de Regressão',
        zorder=1) # zorder=1 coloca a linha atrás dos pontos

plt.xlabel(label_x, fontsize=tamanho_label, color=cor_label)
plt.ylabel(label_y, fontsize=tamanho_label, color=cor_label)
plt.title(f'{titulo_do_grafico} (Sem Legenda de Escala)', fontsize=tamanho_titulo, color=cor_titulo)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

#plt.savefig('scatter_plot_sem_legendade_escala.png', dpi=300)
```
